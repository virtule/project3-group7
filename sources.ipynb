```python
import pandas as pd
import json
```

```python
df = pd.read_csv("csv_sources/sources_flat.csv")
df.head(270)
```

```python
del df["ISO"]
df.head(270)
```

```python
data_list = []
```

```python
for index, row in df.iterrows():
    data_dict = {
        'Country' : row['Country'],
        'Year' : row['Year'],
        'Total' : row['Total'],
        'Coal' : row['Coal'],
        'Oil' : row['Oil'],
        'Gas' : row['Gas'],
        'Cement' : row['Cement'],
        'Flaring' : row['Flaring'],
        'Other' : row['Other'],
        'Per Capita' : row['Per Capita'] 
    }
    data_list.append(data_dict)
```

```python
json_data = json.dumps(data_list)
```

```python
with open("static/js/sources.js", "w") as js_file:
    js_file.write(f"const sources = {json_data};")
```
