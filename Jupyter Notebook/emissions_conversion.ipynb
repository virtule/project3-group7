```python
import pandas as pd
from pandas import json_normalize
import json
import pandas_geojson as pdg
```

```python
emissions_df = pd.read_csv("Resources\GCB2022v27_MtCO2_flat.csv")

countries_df = pd.read_json("Resources\countries.geo.json")

```

```python
emissions_df.head()
```

```python
countries_flattened = json_normalize(countries_df['features'])
```

```python
countries_flattened.head()
```

```python
countries = []

for index, row in countries_flattened.iterrows():
    countries.append(row['properties.name'])

countries
```

```python
dfs = []

for index, row in emissions_df.iterrows():
    if row['Country'] in countries:
        dfs.append(row.to_frame().T)

filtered_countries_df = pd.concat(dfs, ignore_index=True)

```

```python
print(filtered_countries_df)
```

```python
filtered_countries_df = filtered_countries_df.rename(columns={'Country' : 'properties.name'})

filtered_countries_df
```

```python
full_data_df = pd.merge(filtered_countries_df, countries_flattened, on='properties.name', how='inner')

full_data_df
```

```python
full_data_df.rename(columns={'Year' : 'properties.year',
                             'Total' : 'properties.total',
                             'Coal' : 'properties.coal',
                             'Oil' : 'properties.oil',
                             'Gas' : 'properties.gas',
                             'Cement' : 'properties.cement',
                             'Flaring' : 'properties.flaring',
                             'Other' : 'properties.other',
                             'Per Capita' : 'properties.percapita'}, inplace=True)

full_data_df
```

```python
full_data_df = full_data_df.drop('ISO 3166-1 alpha-3', axis=1)

print(full_data_df)
```

```python
full_data_df = full_data_df[['type', 'id', 'properties.name', 'properties.year', 'properties.total', 'properties.coal', 'properties.oil', 
'properties.gas', 'properties.cement', 'properties.flaring', 'properties.other', 'properties.percapita', 'geometry.type', 'geometry.coordinates' ]]

full_data_df = full_data_df.fillna(0)

print(full_data_df)
```

```python
full_data_df.head()
```

```python
full_data_df.to_csv('full_data.csv', index=False)
```

```python
years = [1900, 1910, 1920, 1930, 1940, 1950, 1960, 1970, 1980, 1987, 1990, 1992, 2000, 2005, 2010, 2015, 2020, 2021]

year_highlights_df = full_data_df[full_data_df['properties.year'].isin(years)].copy()

year_highlights_df = pd.DataFrame(year_highlights_df)

year_highlights_df
```

```python
year_highlights_df.rename(columns={
                             'geometry.coordinates' : 'coordinates',
                             'properties.name'	: 'name',
                             'properties.year' : 'year',	
                             'properties.total'	 : 'total',
                             'properties.coal' : 'coal',	
                             'properties.oil' : 'oil',	
                             'properties.gas' : 'gas',	
                             'properties.cement' : 'cement',	
                             'properties.flaring' : 'flaring',	
                             'properties.other' : 'other',	
                             'properties.percapita' : 'percapita'}, inplace=True)
```

```python
print(year_highlights_df)
```

```python
year_highlights_df.to_csv('yearly_emissions_highlights.csv', index=False)
```

```python

geojson = pdg.GeoJSON.from_dataframe(year_highlights_df
                                     ,geometry_type_col='geometry.type'
                                     ,coordinate_col='coordinates'
                                     ,property_col_list=['name','year','total']
                                     )
```

```python
pdg.save_geojson(geojson, 'country_emission_totals.geojson', indent = 3 )
```
